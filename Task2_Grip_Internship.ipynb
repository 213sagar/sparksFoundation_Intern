## **How do yoou find the optimum number of clusters fro K means? How does one determine the value of K**

```python
## method 1: Choosing Randomly 
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y=kmeans.fit_predict(data1.iloc[:,[0,1,2,3]].values)
y
```

```python
sns.pairplot(data1,hue='Species')
```

```python
kmeans.cluster_centers_
```

```python
x=data1.iloc[:,[0,1,2,3]].values
plt.scatter(x[y==0,0],x[y==0,1],s=15,c='red',label='Cluster1')
plt.scatter(x[y==1,0],x[y==1,1],s=15,c='blue',label='Cluster2')
plt.scatter(x[y==2,0],x[y==2,1],s=15,c='green',label='Cluster3')
plt.scatter(x[y==3,0],x[y==3,1],s=15,c='cyan',label='Cluster4')
plt.scatter(x[y==4,0],x[y==4,1],s=15,c='magenta',label='Cluster5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=25,c='yellow',label='Centroid')
plt.legend()
plt.show()
```

```python
## method 2 - Using sum of squares method 
wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

## plotting the reselts onto a line graph,
## allowing us to observe the elbow
plt.plot(range(1,11),wcss)
plt.title('The elbow mwthod')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')## within clusters sum of squares 
plt.show()
```

What is Elbow Methos?

-- The elbow method is a graphical represenattion of finding the optimal 'K' in a K-means clustering. It works by finding WCSS 
(Within-Cluster Sum Of Square ) i.e the sum of the square distance  between points in a cluster and cluster centroid.
